```python
import pandas as pd
```

```python
df = pd.read_csv('male_players.csv')
```

```python
columns_to_drop = ["fifa_update", "player_url","fifa_update_date", "player_face_url", "club_jersey_number", "club_loaned_from","club_contract_valid_until_year", "nation_jersey_number", "work_rate", "body_type", "real_face", "release_clause_eur", "player_tags", "player_traits", "wage_eur", 
                   "ls", "st", "rs", "lw", "lf", "cf",	"rf", "rw", "lam", "cam", "ram", "lm", "lcm", "cm", "rcm", "rm", "lwb", "ldm", "cdm", "rdm", "rwb", "lb", "lcb", "cb", "rcb", "rb", "gk"]
```

```python
for col in columns_to_drop:
    if col in df.columns:
        print(f"Column: {col}")
        print(df[col].unique(), "\n")  # Print unique values to avoid too much output
```

```python
df1 = df.drop(columns=columns_to_drop)
print(df1.head())
```

```python

file_path = 'df1.csv'
df1.to_csv(file_path, index=False)

print(f'DataFrame df1 has been saved as {file_path}')
```
